```python
import pandas as pd

df = pd.read_csv("src/results.csv")
```

```python
df
```

```python
df
# chnage the column names
df.columns = ["Maze size", "Algorithm", "Time mean", "Time std", "Path length mean", "Path length std", "Iterations mean", "Iterations std", "Memory mean", "Memory std", "Visited mean", "Visited std"]
# remove the first and second row
df = df.iloc[2:]
# reset the index
df = df.reset_index(drop=True)
```

```python
df
```

```python
# replace maze size with double values eg. (10, 10) (20, 20) etc.
df["Maze size"] = df["Maze size"].replace({"(5, 5)": "(10, 10)", "(10, 10)": "(20, 20)", "(20, 20)": "(40, 40)", "(50, 50)": "(100, 100)"})
```

```python
# seconds to miliseconds if too small
```

```python
df.info()
# to float
df["Time mean"] = df["Time mean"].astype(float)
df["Time std"] = df["Time std"].astype(float)
df["Path length mean"] = df["Path length mean"].astype(float)
df["Path length std"] = df["Path length std"].astype(float)
df["Iterations mean"] = df["Iterations mean"].astype(float)
df["Iterations std"] = df["Iterations std"].astype(float)
```

```python
df
```

```python
# make every float a string of maximum 8 characters
df["Time mean"] = df["Time mean"].apply(lambda x: "{:.8f}".format(x))
df["Time std"] = df["Time std"].apply(lambda x: "{:.8f}".format(x))
df["Path length mean"] = df["Path length mean"].apply(lambda x: "{:.8f}".format(x))
df["Path length std"] = df["Path length std"].apply(lambda x: "{:.8f}".format(x))
df["Iterations mean"] = df["Iterations mean"].apply(lambda x: "{:.8f}".format(x))
df["Iterations std"] = df["Iterations std"].apply(lambda x: "{:.8f}".format(x))

# maximum 8 characters 
df["Time mean"] = df["Time mean"].apply(lambda x: x[:8])
df["Time std"] = df["Time std"].apply(lambda x: x[:8])
df["Path length mean"] = df["Path length mean"].apply(lambda x: x[:8])
df["Path length std"] = df["Path length std"].apply(lambda x: x[:8])
df["Iterations mean"] = df["Iterations mean"].apply(lambda x: x[:8])
df["Iterations std"] = df["Iterations std"].apply(lambda x: x[:8])

```

```python
df

# # seconds in time to milliseconds
# df["Time mean"] = df["Time mean"].apply(lambda x: float(x)*1000)
# df["Time std"] = df["Time std"].apply(lambda x: float(x)*1000)


df
```

```python
# covert time to miliseconds smae for time std round to 2 decimals
df["Time mean"] = df["Time mean"].apply(lambda x: round(float(x)*1000, 2))
df["Time std"] = df["Time std"].apply(lambda x: round(float(x)*1000, 2))


# round to 2 decimals
df["Path length mean"] = df["Path length mean"].apply(lambda x: round(float(x), 2))
df["Path length std"] = df["Path length std"].apply(lambda x: round(float(x), 2))
df["Iterations mean"] = df["Iterations mean"].apply(lambda x: round(float(x), 2))
df["Iterations std"] = df["Iterations std"].apply(lambda x: round(float(x), 2))

df["Memory mean"] = df["Memory mean"].apply(lambda x: round(float(x), 2))
df["Memory std"] = df["Memory std"].apply(lambda x: round(float(x), 2))

df["Visited mean"] = df["Visited mean"].apply(lambda x: round(float(x), 2))
df["Visited std"] = df["Visited std"].apply(lambda x: round(float(x), 2))
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# set the seaborn style
sns.set()

# get the data
df1 = df[df["Algorithm"] == "MDPPolicyIteration"]
df2 = df[df["Algorithm"] == "MDPValueIteration"]
```

```python


maze_sizes = df["Maze size"].unique()

x = np.arange(len(maze_sizes))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(layout='constrained')

rects1 = ax.bar(x - width/2, df1["Time mean"], width, yerr=df1["Time std"], label='PolicyIteration')
rects2 = ax.bar(x + width/2, df2["Time mean"], width, yerr=df2["Time std"], label='ValueIteration')

# log scale
plt.yscale("log")
# label the bars
ax.set_ylabel('Time (ms)')
ax.set_xlabel('Maze size')
# set x ticks
ax.set_xticks(x)
ax.set_xticklabels(maze_sizes)

plt.legend(loc = "upper left")
# sae the figure
plt.savefig("Figures/timemdp.png")

plt.show()

# same for iterations
fig, ax = plt.subplots(layout='constrained')

rects1 = ax.bar(x - width/2, df1["Iterations mean"], width, yerr=df1["Iterations std"], label='PolicyIteration',)
rects2 = ax.bar(x + width/2, df2["Iterations mean"], width, yerr=df2["Iterations std"], label='ValueIteration')

# label the bars
ax.set_ylabel('Iterations')
ax.set_xlabel('Maze size')
# set x ticks
ax.set_xticks(x)
ax.set_xticklabels(maze_sizes)

plt.legend(loc = "upper left")
# sae the figure
plt.savefig("Figures/iterations.png")
plt.show()
```
