```python
import pandas as pd

df = pd.read_csv("src/results.csv")
```

```python
df
# chnage the column names
df.columns = ["Maze size", "Algorithm", "Time mean", "Time std", "Path length mean", "Path length std", "Iterations mean", "Iterations std"]
# remove the first and second row
df = df.iloc[2:]
# reset the index
df = df.reset_index(drop=True)
```

```python
df.info()
# to float
df["Time mean"] = df["Time mean"].astype(float)
df["Time std"] = df["Time std"].astype(float)
df["Path length mean"] = df["Path length mean"].astype(float)
df["Path length std"] = df["Path length std"].astype(float)
df["Iterations mean"] = df["Iterations mean"].astype(float)
df["Iterations std"] = df["Iterations std"].astype(float)
```
